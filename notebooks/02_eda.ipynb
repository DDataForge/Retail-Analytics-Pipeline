```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
retail = pd.read_csv("../data/processed/retail_clean.csv")
train = pd.read_csv("../data/processed/train_clean.csv")

# Convert dates
retail['transaction_date'] = pd.to_datetime(retail['transaction_date'])
train['order_date'] = pd.to_datetime(train['order_date'])
```

```python
#1. SALES TREND (TRAIN)
# -----------------------------
sales_trend = train.groupby(train['order_date'].dt.to_period('M'))['sales'].sum()
sales_trend.index = sales_trend.index.astype(str)

plt.figure()
sales_trend.plot(title="Monthly Sales Trend (Train Dataset)")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
#2. PAYMENT METHODS (RETAIL)
# -----------------------------
plt.figure()
sns.countplot(x='payment_method', data=retail)
plt.title("Payment Method Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
#3. CATEGORY PERFORMANCE (TRAIN)
# -----------------------------
plt.figure()
sns.barplot(x='category', y='sales', data=train)
plt.title("Sales by Category (Train Dataset)")
plt.xticks(rotation=45)
plt.show()
```

```python
#4. TOP 10 PRODUCTS
# -----------------------------
top_products = train.groupby('product_name')['sales'].sum().nlargest(10)

plt.figure()
top_products.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
#5. CUSTOMER SEGMENTATION (RETAIL)
# -----------------------------
customer_sales = retail.groupby('customer_id')['total_spent'].sum().reset_index()

plt.figure()
sns.histplot(customer_sales['total_spent'], bins=30, kde=True)
plt.title("Customer Spending Distribution")
plt.xlabel("Total Spent")
plt.show()
```

```python
#6. SALES BY REGION (TRAIN)
# -----------------------------
plt.figure()
sns.barplot(x='region', y='sales', data=train)
plt.title("Sales by Region")
plt.xticks(rotation=45)
plt.show()
```

```python
#7. SALES BY CATEGORY (RETAIL)
# -----------------------------
plt.figure()
sns.barplot(x='category', y='total_spent', data=retail)
plt.title("Retail Sales by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
#8. PROFIT ANALYSIS (TRAIN ONLY)
# -----------------------------
if 'profit' in train.columns:
    plt.figure()
    sns.barplot(x='category', y='profit', data=train)
    plt.title("Profit by Category")
    plt.xticks(rotation=45)
    plt.show()

    plt.figure()
    sns.histplot(train['profit'], bins=30, kde=True)
    plt.title("Profit Distribution")
    plt.show()
```
